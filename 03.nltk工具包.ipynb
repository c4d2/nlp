#### 分词

```python
from nltk.tokenize import word_tokenize
from nltk.text import Text
```

```python
str = "Today's weather is good,very windy and sunny,we have no classes in the afternoon,We have to play basketball tomorrow"
```

```python
tokens = word_tokenize(str)
```

```python
tokens = [word.lower() for word in tokens]
tokens
```

#### 创建Text对象，方便后续操作

```python
t = Text(tokens)
```

```python
# 统计tokens中的文字数量
t.count('good')
```

```python
# 统计出现good这个词的位置索引
t.index('good')
```

```python
# 将前八个字符的数量统计起来并绘制成表格
t.plot(8)
```

#### 停用词 类似于the这种词，没有情感色彩，比较无聊的词语

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
# 停用词所支持的语言种类
stopwords.fileids()
```

```python
stopwords.raw('english').replace('\n',' ')
```

```python
words = [word.lower() for word in tokens]
words
```

```python
words = set(words)
```

```python
# 查看一下停用词和我们当前文本的交集，也就是说看一下当前文本有哪些是停用词并打印出来
words.intersection(set(stopwords.words('english')))
```

#### 过滤停用词

```python
[word for word in words if(word not in stopwords.words('english'))]
```
