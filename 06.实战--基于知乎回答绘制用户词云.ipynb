```python
import jieba
# 读取文本中的内容
def read_file(filename):
    with open(filename,'r') as file:
        return file.read()

text = read_file('./data/ljc.txt')
text = text.replace('\n',' ')
text_list = list(jieba.cut(text))
text_list
```

```python
import jieba.analyse
tags = jieba.analyse.extract_tags(text,withWeight=True)
for word,weight in tags:
#     权重值代表每个词的重要程度
    print (word,weight)
```

```python
# 加载停用词
stopwords = read_file('./data/stopwords/cn_stopwords.txt')
# 分割文档中的停用词
stopwors = stopwords.split("\n")
```

```python
# 去除读取的文本中的停用词
result = [word for word in text_list if word not in stopwords]
result
```

```python
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt

word_counts = Counter(result)
word_freq = dict(word_counts)

print(word_freq)
# 创建词云对象
wc = WordCloud(font_path="./data/font/AdobeHeitiStd-Regular.otf",background_color="white", width=800, height=400)

# 生成词云图
wc.generate_from_frequencies(word_freq)

# 展示词云图
plt.imshow(wc)
plt.axis("off")
plt.show()
```
